```python
import numpy as np
import matplotlib.pyplot as plt

# Define the file path
file_path = "horizons/etk_output_file.dat"

# Define the shape of the 4D array
# Replace these dimensions with the actual sizes used in your C++ code
dim1 = 21  # Example size for the first dimension
dim2 = 21  # Example size for the second dimension
dim3 = 21  # Example size for the third dimension
dim4 = 16  # Example size for the fourth dimension

# Define the data type (assumed to be double or float)
# Use np.float64 for Real if Real is a double, or np.float32 if it's a float
data_type = np.float64  # Adjust according to the `Real` type in your C++ code

# Calculate the total number of elements
count = dim1 * dim2 * dim3 * dim4

# Read the binary file
try:
    with open(file_path, "rb") as file:
        # Read the binary data and convert it to a 1D numpy array
        data = np.fromfile(file, dtype=data_type, count=count)

    # Reshape the data to a 4D array
    data_reshaped = data.reshape((dim1, dim2, dim3, dim4))

    # Print or process the data as needed
    print("Data read from the file:")
    print(data_reshaped)

except FileNotFoundError:
    print(f"Error: File '{file_path}' not found.")
except Exception as e:
    print(f"Error reading the file: {e}")
```
